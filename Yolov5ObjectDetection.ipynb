```python
import os
import cv2
import matplotlib.pyplot as plt
import yaml

# Paths
path = '/home/db21160/Desktop/yoloCode/yolov5/runs/detect/exp/'
label_path = os.path.join(path, 'labels')
output_path = os.path.join(path, 'boxed_images')
os.makedirs(output_path, exist_ok=True)

# Load class names from the data.yaml file
with open('/home/db21160/Desktop/yoloCode/yolov5/data/coco128.yaml', 'r') as file:
    data = yaml.safe_load(file)
class_names = data['names']

for file in os.listdir(path):
    if not file.lower().endswith(('.jpg', '.png', '.jpeg')):
        continue  # skip non-image files

    print('---------------', file, '-------------------------')
    labelfile = os.path.join(label_path, '.'.join(file.split('.')[:-1]) + '.txt')
    print('Label file:', labelfile)

    img_path = os.path.join(path, file)
    img = cv2.imread(img_path)

    if img is None:
        print('Image not found or unreadable:', img_path)
        continue

    height, width, _ = img.shape

    if os.path.exists(labelfile):
        with open(labelfile, 'r') as myfile:
            lines = myfile.readlines()

        for line in lines:
            parts = line.strip().split()
            if len(parts) < 5:
                continue  # skip malformed lines

            cls, x_center, y_center, w, h = map(float, parts[:5])

            # Convert normalized coords to absolute pixel values
            x_center *= width
            y_center *= height
            w *= width
            h *= height

            # Top-left and bottom-right points
            x1 = int(x_center - w / 2)
            y1 = int(y_center - h / 2)
            x2 = int(x_center + w / 2)
            y2 = int(y_center + h / 2)

            # Prepare label text
            label_text = class_names[int(cls)]  # Class name instead of number
            font_scale = 1.2  # bigger text
            font_thickness = 3  # thicker text
            font = cv2.FONT_HERSHEY_SIMPLEX

            # Calculate text size for background
            (text_width, text_height), baseline = cv2.getTextSize(label_text, font, font_scale, font_thickness)
            cv2.rectangle(img, (x1, y1 - text_height - 10), (x1 + text_width, y1), (255, 255, 255), -1)  # white background for text

            # Draw the text on top of white box
            cv2.putText(img, label_text, (x1, y1 - 5), font, font_scale, (0, 0, 0), font_thickness)

            box_thickness = 5  # Increased box thickness
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), box_thickness)

        # Save boxed image
        boxed_img_path = os.path.join(output_path, file)
        cv2.imwrite(boxed_img_path, img)
        print(f"Saved with boxes: {boxed_img_path}")

        # Display
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img_rgb)
        plt.axis('off')
        plt.title(f"Boxed: {file}")
        plt.show()

    else:
        print('No detection label found for this image.')
```
